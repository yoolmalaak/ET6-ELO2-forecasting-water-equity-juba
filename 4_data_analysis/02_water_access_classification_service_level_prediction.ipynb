## Supervised Learning Framework for School Potable Water Prediction

```python
# Notebook 2: School Potable Water Classification ‚Äî Service Level Prediction

# imports
import os
import re
import json
import time
import joblib
import warnings
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import (
    train_test_split,
    StratifiedKFold,
    cross_validate,
    RandomizedSearchCV,
)
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import (
    VarianceThreshold,
    SelectKBest,
    mutual_info_classif,
)
from sklearn.inspection import permutation_importance, PartialDependenceDisplay

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    RocCurveDisplay,
    PrecisionRecallDisplay,
)

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")

# Try imblearn.SMOTE
try:
    from imblearn.over_sampling import SMOTE
    from imblearn.pipeline import Pipeline as ImbPipeline

    SMOTE_AVAILABLE = True
    print("‚úî imblearn.SMOTE available ‚Äî will use inside pipelines.")
except Exception:
    from sklearn.pipeline import Pipeline as ImbPipeline  # fallback naming

    SMOTE_AVAILABLE = False
    print(" imblearn.SMOTE not available ‚Äî pipelines will not include SMOTE.")

# ============================================================
# UNIVERSAL CHART-SAVING FUNCTION
# Saves all charts to 5_communication/charts/
# ============================================================
CHART_DIR = "../5_communication/charts"
os.makedirs(CHART_DIR, exist_ok=True)


def save_fig(name, fig=None, dpi=150):
    if fig is None:
        fig = plt.gcf()
    path = os.path.join(CHART_DIR, f"{name}.png")
    fig.savefig(path, dpi=dpi, bbox_inches="tight")
    print(f"üìÅ Saved chart: {path}")


# -------------------------
# CONFIG
# -------------------------
RANDOM_STATE = 42
DATA_PATH = "../2_data_preparation/final_cleaned_dataset.csv"  # change if needed
SAVE_DIR = "./notebook2_outputs"
os.makedirs(SAVE_DIR, exist_ok=True)

FINAL_TARGET = "school_has_potable_water_numeric"
MIN_ROWS = 5

LEAKAGE_COLS = [
    # direct or derived potable indicators EXCEPT the FINAL TARGET
    "school_has_potable_water",
    "school_has_potable_water_clean",
    # behaviors revealing potable access
    "school_drinking_frequency",
    "school_has_handwashing_facilities",
    "school_handwashing_frequency",
    # indirect potable indicators
    "school_equitable_access",
    "school_sleep_without_bathing",
    "school_sleep_without_bathing_frequency",
    "school_water_source",
    "school_hours_water_available_numeric",
]


# load data
df = pd.read_csv(DATA_PATH)
print("Initial shape:", df.shape)
if df.shape[0] == 0:
    raise SystemExit("Dataset is empty. Aborting.")


# reconstruct target
def _map_text_to_binary(s):
    if pd.isna(s):
        return np.nan
    s = str(s).strip().lower()
    s = re.sub(r"[^a-z0-9\s]+", " ", s)
    if s == "":
        return np.nan
    yes_tokens = (
        "yes",
        "y",
        "true",
        "available",
        "functional",
        "potable",
        "safe",
        "improved",
        "tap",
        "piped",
    )
    no_tokens = (
        "no",
        "n",
        "false",
        "not available",
        "not functional",
        "unavailable",
        "unsafe",
        "unimproved",
    )
    for t in yes_tokens:
        if t in s:
            return 1
    for t in no_tokens:
        if t in s:
            return 0
    if "sometimes" in s or "partial" in s or "intermittent" in s or "limited" in s:
        return np.nan
    if s.isdigit():
        v = int(s)
        if v in (0, 1):
            return v
    return np.nan


# Prefer clean column if present, else raw
if (
    "school_has_potable_water_clean" in df.columns
    and df["school_has_potable_water_clean"].notna().any()
):
    print("Reconstructing target from 'school_has_potable_water_clean'...")
    df[FINAL_TARGET] = df["school_has_potable_water_clean"].apply(_map_text_to_binary)
elif (
    "school_has_potable_water" in df.columns
    and df["school_has_potable_water"].notna().any()
):
    print("Reconstructing target from 'school_has_potable_water'...")
    df[FINAL_TARGET] = df["school_has_potable_water"].apply(_map_text_to_binary)
else:
    raise SystemExit("No potable-water column found to reconstruct target.")

print("Target counts (before dropna):")
print(df[FINAL_TARGET].value_counts(dropna=False))

# Drop rows with missing target
df = df.dropna(subset=[FINAL_TARGET]).copy()
df[FINAL_TARGET] = df[FINAL_TARGET].astype(int)
print("Target counts (after dropna):")
print(df[FINAL_TARGET].value_counts())


# drop leakage columns
cols_to_drop = [c for c in LEAKAGE_COLS if c in df.columns]
if cols_to_drop:
    print("Dropping leakage columns:", cols_to_drop)
    df = df.drop(columns=cols_to_drop)
else:
    print("No leakage columns found to drop.")

# confirm FINAL_TARGET still present
if FINAL_TARGET not in df.columns:
    raise SystemExit(f"{FINAL_TARGET} missing after processing. Aborting.")

# define X, y and safe split
X = df.drop(columns=[FINAL_TARGET])
y = df[FINAL_TARGET]

if df.shape[0] < MIN_ROWS:
    raise SystemExit(
        f"Not enough rows ({df.shape[0]}) to model. Need >= {MIN_ROWS} rows."
    )

print("Rows usable for modeling:", df.shape[0])
print("Target distribution:", y.value_counts().to_dict())

stratify_arg = y if y.value_counts().min() >= 2 else None

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=RANDOM_STATE, stratify=stratify_arg, shuffle=True
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)


# preprocessor factory (returns a fresh preprocessor)
def make_preprocessor(numeric_cols, categorical_cols):
    numeric_transformer = Pipeline(
        [("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())]
    )
    categorical_transformer = Pipeline(
        [
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
        ]
    )
    preproc = ColumnTransformer(
        [
            ("num", numeric_transformer, numeric_cols),
            ("cat", categorical_transformer, categorical_cols),
        ],
        remainder="drop",
    )
    return preproc


# detect column types on training set
numeric_cols = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X_train.select_dtypes(
    include=["object", "bool", "category"]
).columns.tolist()
print(f"Numeric cols: {len(numeric_cols)}; Categorical cols: {len(categorical_cols)}")

# Create a temporary preprocessor to compute feature count and k_best (this is safe;
# pipelines used inside CV will build fresh preprocessors per fold)
temp_preproc = make_preprocessor(numeric_cols, categorical_cols)
temp_preproc.fit(X_train)  # fit only to compute engineered feature names for k_best
try:
    engineered_names = temp_preproc.get_feature_names_out()
except Exception:
    engineered_names = np.array(
        [f"f_{i}" for i in range(temp_preproc.transform(X_train.iloc[:1]).shape[1])]
    )
print("Engineered feature count:", len(engineered_names))

k_best = min(30, len(engineered_names))
print("Selecting top k =", k_best)


# pipeline builder
def build_pipeline(model, use_smote=SMOTE_AVAILABLE, k_best_local=k_best):
    preproc = make_preprocessor(numeric_cols, categorical_cols)  # fresh instance
    steps = [
        ("pre", preproc),
        ("var", VarianceThreshold(0.0)),
        ("select", SelectKBest(score_func=mutual_info_classif, k=k_best_local)),
    ]
    if use_smote:
        steps.append(("smote", SMOTE(random_state=RANDOM_STATE)))
    steps.append(("clf", model))
    return ImbPipeline(steps)


# model baselines
models = {
    "Logistic": LogisticRegression(
        solver="liblinear", class_weight="balanced", max_iter=500
    ),
    "RandomForest": RandomForestClassifier(
        n_estimators=300, class_weight="balanced", random_state=RANDOM_STATE
    ),
    "GradientBoosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric="logloss",
        random_state=RANDOM_STATE,
    ),
}

scoring = {
    "accuracy": "accuracy",
    "precision_macro": "precision_macro",
    "recall_macro": "recall_macro",
    "f1_macro": "f1_macro",
    "roc_auc": "roc_auc",
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)


# cross-validation
cv_results = {}
print(
    "\nRunning cross-validation (fresh preprocessor inside pipeline for each fold)..."
)
for name, model in models.items():
    pipe = build_pipeline(model)
    try:
        scores = cross_validate(
            pipe,
            X_train,
            y_train,
            scoring=scoring,
            cv=cv,
            n_jobs=-1,
            return_train_score=False,
        )
    except Exception as e:
        print("CV parallel failed, retrying single-threaded:", e)
        scores = cross_validate(
            pipe,
            X_train,
            y_train,
            scoring=scoring,
            cv=cv,
            n_jobs=1,
            return_train_score=False,
        )
    summary = {
        metric: (np.mean(scores[f"test_{metric}"]), np.std(scores[f"test_{metric}"]))
        for metric in scoring
    }
    cv_results[name] = summary
    print(f"\n{name} CV summary:")
    for metric, (m_mean, m_std) in summary.items():
        print(f"  {metric}: {m_mean:.4f} ¬± {m_std:.4f}")

# Summarize CV results into DataFrame
cv_rows = []
for name, metrics in cv_results.items():
    row = {"model": name}
    for metric, (meanv, stdv) in metrics.items():
        row[f"{metric}_mean"] = meanv
        row[f"{metric}_std"] = stdv
    cv_rows.append(row)
cv_summary_df = (
    pd.DataFrame(cv_rows)
    .sort_values("roc_auc_mean", ascending=False)
    .reset_index(drop=True)
)
print("\nCV summary (sorted):")
display(cv_summary_df)


# -------------------------
# HYPERPARAMETER TUNING (RandomizedSearchCV for RF & XGB)
# -------------------------
print("\nRunning randomized hyperparameter tuning (RandomForest, XGBoost)...")
n_iter = 20

# RF param space
rf_param_dist = {
    "clf__n_estimators": [100, 200, 300, 500],
    "clf__max_depth": [None, 5, 10, 20],
    "clf__min_samples_split": [2, 5, 10],
    "clf__min_samples_leaf": [1, 2, 4],
    "clf__max_features": ["sqrt", "log2", None],
}

# XGB param space
xgb_param_dist = {
    "clf__n_estimators": [100, 200, 300],
    "clf__max_depth": [3, 4, 6],
    "clf__learning_rate": [0.01, 0.03, 0.05, 0.1],
    "clf__subsample": [0.6, 0.8, 0.9, 1.0],
    "clf__colsample_bytree": [0.6, 0.8, 1.0],
}

tune_results = {}


def run_random_search(base_model, param_dist, name, n_iter_local=n_iter):
    pipe = build_pipeline(base_model)
    rs = RandomizedSearchCV(
        pipe,
        param_distributions=param_dist,
        n_iter=n_iter_local,
        scoring="roc_auc",
        cv=cv,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        verbose=0,
    )
    try:
        rs.fit(X_train, y_train)
    except Exception as e:
        print(f"RandomizedSearchCV for {name} failed:", e)
        return None
    print(f"{name} best ROC-AUC: {rs.best_score_:.4f}")
    return rs


rf_rs = run_random_search(
    RandomForestClassifier(random_state=RANDOM_STATE, class_weight="balanced"),
    rf_param_dist,
    "RandomForest",
)
xgb_rs = run_random_search(
    XGBClassifier(
        random_state=RANDOM_STATE, use_label_encoder=False, eval_metric="logloss"
    ),
    xgb_param_dist,
    "XGBoost",
)

if rf_rs is not None:
    tune_results["RandomForest"] = {
        "best_estimator": rf_rs.best_estimator_,
        "best_score": rf_rs.best_score_,
        "best_params": rf_rs.best_params_,
    }
if xgb_rs is not None:
    tune_results["XGBoost"] = {
        "best_estimator": xgb_rs.best_estimator_,
        "best_score": xgb_rs.best_score_,
        "best_params": xgb_rs.best_params_,
    }

# Update cv summary with tuned results
if "RandomForest" in tune_results:
    cv_summary_df.loc[cv_summary_df.model == "RandomForest", "roc_auc_mean"] = (
        tune_results["RandomForest"]["best_score"]
    )
if "XGBoost" in tune_results:
    cv_summary_df.loc[cv_summary_df.model == "XGBoost", "roc_auc_mean"] = tune_results[
        "XGBoost"
    ]["best_score"]

cv_summary_df = cv_summary_df.sort_values("roc_auc_mean", ascending=False).reset_index(
    drop=True
)
print("\nUpdated CV summary (with tuned results if available):")
display(cv_summary_df)

# select best model
best_model_name = cv_summary_df.loc[0, "model"]
print("Best model selected by CV ROC-AUC:", best_model_name)

# prefer tuned estimator if available
if best_model_name in tune_results:
    final_pipe = tune_results[best_model_name]["best_estimator"]
else:
    final_pipe = build_pipeline(models[best_model_name])

# final_pipe is ready to be fit on full training set
print("\nFitting final pipeline on full training set...")
start_time = time.time()
final_pipe.fit(X_train, y_train)
print("Fitted in %.1f seconds" % (time.time() - start_time))

# holdout evaluation

y_pred = final_pipe.predict(X_test)
y_proba = (
    final_pipe.predict_proba(X_test)[:, 1]
    if hasattr(final_pipe, "predict_proba")
    else None
)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
auc = roc_auc_score(y_test, y_proba) if y_proba is not None else np.nan

print("\nHoldout results:")
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1:        {f1:.4f}")
print(f"ROC-AUC:   {auc if not np.isnan(auc) else 'n/a'}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# ROC and Precision‚ÄìRecall curves
if y_proba is not None:
    # ROC Curve
    plt.figure()
    RocCurveDisplay.from_predictions(y_test, y_proba)
    plt.title("ROC Curve")
    save_fig("notebook2_roc_curve")
    plt.show()

    # Precision‚ÄìRecall Curve
    plt.figure()
    PrecisionRecallDisplay.from_predictions(y_test, y_proba)
    plt.title("Precision-Recall Curve")
    save_fig("notebook2_precision_recall_curve")
    plt.show()

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix ‚Äî {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
save_fig("notebook2_confusion_matrix")
plt.show()


# permutation importance, PDPs, error analysis
print("\nComputing permutation importance (may take a few seconds)...")
pi = permutation_importance(
    final_pipe,
    X_test,
    y_test,
    n_repeats=30,
    scoring="roc_auc",
    random_state=RANDOM_STATE,
)

# Extract feature names AFTER preproc -> var -> select
# Step 1: preprocessor engineered names (from temp_preproc fitted earlier)
pre_names = engineered_names  # from temp_preproc.get_feature_names_out()
# Step 2: apply var mask
var_mask = final_pipe.named_steps["var"].get_support()
var_names = np.array(pre_names)[var_mask]
# Step 3: apply select mask
sel_mask = final_pipe.named_steps["select"].get_support()
selected_feature_names = list(np.array(var_names)[sel_mask])

pi_df = (
    pd.DataFrame(
        {
            "feature": selected_feature_names,
            "importance_mean": pi.importances_mean[: len(selected_feature_names)],
            "importance_std": pi.importances_std[: len(selected_feature_names)],
        }
    )
    .sort_values("importance_mean", ascending=False)
    .reset_index(drop=True)
)

plt.figure(figsize=(8, 6))
sns.barplot(data=pi_df.head(20), x="importance_mean", y="feature")
plt.title("Top 20 Permutation Importances")
plt.tight_layout()
save_fig("notebook2_permutation_importance")
plt.show()


pi_df.to_csv(os.path.join(SAVE_DIR, "permutation_importance.csv"), index=False)

# partial dependence plots (PDPs) for top 3 features
top_feats = pi_df.head(3)["feature"].tolist()
print("\nGenerating PDPs for:", top_feats)
for feat in top_feats:
    try:
        PartialDependenceDisplay.from_estimator(
            final_pipe, X_train, features=[feat], feature_names=X_train.columns
        )
        plt.title(f"PDP: {feat}")
        plt.show()
    except Exception as e:
        print("Could not generate PDP for", feat, ":", e)


# error analysis

pred_df = X_test.copy().reset_index(drop=True)
pred_df["actual"] = y_test.reset_index(drop=True)
pred_df["pred"] = y_pred
if y_proba is not None:
    pred_df["prob"] = y_proba

fp = pred_df[(pred_df["actual"] == 0) & (pred_df["pred"] == 1)]
fn = pred_df[(pred_df["actual"] == 1) & (pred_df["pred"] == 0)]
print("\nFalse positives:", len(fp), "False negatives:", len(fn))
if len(fp) > 0:
    display(fp.head())
if len(fn) > 0:
    display(fn.head())

# save final pipeline, predictions, cv summary, pi
pipeline_path = os.path.join(SAVE_DIR, f"final_pipeline_{best_model_name}.joblib")
joblib.dump(final_pipe, pipeline_path)
print("Saved pipeline to:", pipeline_path)

preds_path = os.path.join(SAVE_DIR, "holdout_predictions.csv")
pred_df.to_csv(preds_path, index=False)
print("Saved holdout predictions to:", preds_path)

cv_summary_path = os.path.join(SAVE_DIR, "cv_summary.csv")
cv_summary_df.to_csv(cv_summary_path, index=False)
print("Saved CV summary to:", cv_summary_path)

pi_path = os.path.join(SAVE_DIR, "permutation_importance.csv")
print("Saved permutation importance to:", pi_path)


# inference function


def predict_school_potable_water(new_df, pipeline_path=pipeline_path):
    """
    new_df: raw dataframe with same schema as training X (no preprocessing)
    pipeline_path: path to saved final pipeline
    Returns new_df with 'pred' and 'prob' columns appended.
    """
    pl = joblib.load(pipeline_path)
    preds = pl.predict(new_df)
    out = pd.DataFrame({"pred": preds})
    if hasattr(pl, "predict_proba"):
        out["prob"] = pl.predict_proba(new_df)[:, 1]
    return pd.concat([new_df.reset_index(drop=True), out], axis=1)


report_html = os.path.join(SAVE_DIR, "notebook2_report.html")
html = f"""
<html>
<head><meta charset="utf-8"><title>Notebook2 Report</title></head><body>
<h1>Notebook 2 ‚Äî School Potable Water Classification</h1>
<p>Generated: {datetime.utcnow().isoformat()} UTC</p>
<p>Rows used: {df.shape[0]}</p>
<p>Target distribution: {df[FINAL_TARGET].value_counts().to_dict()}</p>
<h2>CV Summary</h2>
{cv_summary_df.to_html(index=False, float_format="%.4f")}
<h2>Holdout Metrics</h2>
<ul>
<li>Accuracy: {acc:.4f}</li>
<li>Precision: {prec:.4f}</li>
<li>Recall: {rec:.4f}</li>
<li>F1: {f1:.4f}</li>
<li>ROC-AUC: {auc if not np.isnan(auc) else "n/a"}</li>
</ul>
<h2>Top Permutation Importances</h2>
<pre>{json.dumps(pi_df.head(10).to_dict(orient="records"), indent=2)}</pre>
</body></html>
"""
with open(report_html, "w", encoding="utf-8") as f:
    f.write(html)
print("Saved simple HTML report to:", report_html)

print("\nNotebook 2 completed successfully.")
```
