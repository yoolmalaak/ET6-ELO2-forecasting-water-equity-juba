## Household Data Exploration

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import os


# Path where charts will be saved
CHART_DIR = "./charts"  # because you're already inside 3_data_exploration
os.makedirs(CHART_DIR, exist_ok=True)


def save_fig(name, fig=None, dpi=150):
    """
    Saves a matplotlib figure to the 3_data_exploration/charts folder.

    Parameters:
        name (str): File name without extension.
        fig (matplotlib.figure.Figure, optional): Figure to save.
            If None, the current active figure is used.
        dpi (int): Image resolution.
    """
    if fig is None:
        fig = plt.gcf()

    path = os.path.join(CHART_DIR, f"{name}.png")
    fig.savefig(path, dpi=dpi, bbox_inches="tight")
    print(f"üìÅ Saved chart to: {path}")


# --------------------------------------------------------
# LOAD DATA
# --------------------------------------------------------
dataset = pd.read_csv("../2_data_preparation/final_cleaned_dataset.csv")

# --------------------------------------------------------
# QUICK INSPECTION
# --------------------------------------------------------
print("\n--- Dataset Shape ---")
print(dataset.shape)

print("\n--- First 5 Rows ---")
display(dataset.head())

print("\n--- Column Info ---")
print(dataset.info())

print("\n--- Summary Statistics ---")
display(dataset.describe(include="all"))

# --------------------------------------------------------
# HOUSEHOLD VARIABLES (EXACT COLUMN NAMES)
# --------------------------------------------------------
age_col = "age_group"
hh_size_col = "household_size"
source_col = "main_drinking_water"
shortage_freq_col = "household_shortage_frequency"
distance_col = "distance_to_water_source"

gender_col = "gender"
role_col = "role_in_household"
education_col = "household_head_education"
income_col = "primary_income"
litres_col = "litres_per_person_per_day"


# PLOTTING FUNCTION
# ============================
# FIXED PLOTTING FUNCTIONS
# ============================


def plot_bar(column, title, xlabel="", rotate=45):
    plt.figure(figsize=(8, 4))
    dataset[column].value_counts().plot(kind="bar")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Count")
    plt.xticks(rotation=rotate)
    plt.tight_layout()

    # SAVE CHART
    safe_name = re.sub(r"[^a-zA-Z0-9_]+", "_", title.lower())
    save_fig(safe_name)

    plt.show()


def plot_hist(column, title, xlabel=""):
    plt.figure(figsize=(8, 4))
    dataset[column].hist()
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.tight_layout()

    # SAVE CHART
    safe_name = re.sub(r"[^a-zA-Z0-9_]+", "_", title.lower())
    save_fig(safe_name)

    plt.show()


# EXPLORATION & VISUALIZATION


# Age group
plot_bar(age_col, "Age Group Distribution", "Age Group", rotate=0)

#  Gender
plot_bar(gender_col, "Gender Distribution", "Gender", rotate=0)

# Role in household
plot_bar(role_col, "Role in Household", "Role")

# Education level
plot_bar(education_col, "Household Head Education Level", "Education Level")

# Primary income source
plot_bar(income_col, "Primary Source of Household Income", "Income Source")

# Daily litres per person
plot_hist(litres_col, "Daily Water Consumption per Person", "Litres per Person per Day")

# Household size
plot_hist(hh_size_col, "Household Size Distribution", "Household Size")

# Main drinking water source
plot_bar(source_col, "Main Drinking Water Source", "Source")

# Water shortage frequency
plot_bar(
    shortage_freq_col, "Water Shortage Frequency (Last 12 Months)", "Shortage Frequency"
)


# Distance to water source
plot_bar(distance_col, "Distance to Water Source (One Way)", "Distance")
```

## SCHOOL EXPLORATION

```python
# Load dataset
school_data = dataset.copy()

print("Dataset loaded successfully!")
print(school_data.head())

# -------------------------------
# CLEANING FUNCTIONS
# -------------------------------


def clean_population(value):
    """Convert values like '1,001‚Äì2,000', 'Less than 50', 'More than 500' into numeric midpoints."""
    if pd.isna(value):
        return np.nan

    value = str(value).strip()

    # Standardize dash symbols
    for dash in ["‚Äì", "‚Äî"]:
        value = value.replace(dash, "-")

    # Remove commas and spaces
    cleaned = value.replace(",", "").replace(" ", "").lower()

    # Case: Less than X
    if cleaned.startswith("lessthan"):
        try:
         num = int(cleaned.replace("lessthan", ""))
         return num / 2
        except Exception:
         return np.nan


    # Case: More than X
    if cleaned.startswith("morethan"):

        try:
         num = int(cleaned.replace("morethan", ""))
         return num + 100
        except Exception:
         return np.nan
    

    # Case: Range X-Y
    match = re.match(r"(\d+)-(\d+)", cleaned)
    if match:
        low = int(match.group(1))
        high = int(match.group(2))
        return (low + high) / 2

    # Case: Single number
    
    try:
        return float(cleaned)
    except Exception:
        return np.nan



# -------------------------------
# APPLY CLEANING
# -------------------------------

# Fix student + staff population
school_data["school_student_population_numeric"] = school_data[
    "school_student_population"
].apply(clean_population)
school_data["school_staff_population_numeric"] = school_data[
    "school_staff_population"
].apply(clean_population)

# Fix monthly water cost and remove unrealistic outliers
school_data["school_monthly_water_cost_ssp"] = pd.to_numeric(
    school_data["school_monthly_water_cost_ssp"], errors="coerce"
)
# Remove values above 5 million SSP (likely incorrect/scientific notation)
school_data.loc[
    school_data["school_monthly_water_cost_ssp"] > 5_000_000,
    "school_monthly_water_cost_ssp",
] = np.nan

# Convert to USD (1 USD = 4500 SSP)
school_data["school_monthly_water_cost_usd"] = (
    school_data["school_monthly_water_cost_ssp"] / 4500
)

# Convert boolean categories to numeric
school_data["school_has_potable_water_numeric"] = school_data[
    "school_has_potable_water_clean"
].map({"Yes": 1, "No": 0})

print("\nCleaning completed!\n")

# -------------------------------
# SCHOOL NUMERICAL EXPLORATION
# -------------------------------

school_numeric_cols = [
    "school_student_population_numeric",
    "school_staff_population_numeric",
    "school_monthly_water_cost_ssp",
    "school_monthly_water_cost_usd",
    "school_hours_water_available_numeric",
    "school_drinking_frequency",
    "school_handwashing_frequency_numeric",
    "school_sleep_without_bathing_numeric",
    "school_sleep_without_bathing_frequency_numeric",
    "school_has_potable_water_numeric",
]

# Keep only those that exist
school_numeric_cols = [c for c in school_numeric_cols if c in school_data.columns]

school_numeric_df = school_data[school_numeric_cols]

print("SCHOOL NUMERICAL SUMMARY:\n")
print(school_numeric_df.describe().T)


# -------------------------------
# VISUALIZATIONS
# -------------------------------

# School Type Distribution
plt.figure(figsize=(7, 4))
school_data["school_type"].value_counts().plot(kind="bar", color="steelblue")
plt.title("Distribution of School Types")
plt.ylabel("Count")
save_fig("school_type_distribution")
plt.show()

# Do Schools Have Potable Water?
plt.figure(figsize=(7, 4))
school_data["school_has_potable_water_clean"].value_counts().plot(
    kind="bar", color="orange"
)
plt.title("Schools With Potable Water")
plt.ylabel("Count")
save_fig("schools_with_potable_water")
plt.show()

# Water Source at Schools
plt.figure(figsize=(8, 4))
school_data["school_water_source"].value_counts().plot(kind="bar", color="green")
plt.title("Main Water Source at Schools")
plt.ylabel("Count")
plt.xticks(rotation=45)
save_fig("main_water_source_at_schools")
plt.show()

# Hours of Water Available Per Day
plt.figure(figsize=(7, 4))
plt.hist(
    school_data["school_hours_water_available_numeric"].dropna(),
    bins=10,
    color="purple",
)
plt.title("Daily Hours of Water Availability at Schools")
plt.xlabel("Hours")
plt.ylabel("Frequency")
save_fig("school_hours_water_available")
plt.show()

# üîß Force numeric conversion for ALL numeric columns
school_numeric_df = school_numeric_df.apply(pd.to_numeric, errors="coerce")

# Correlation Heatmap (now guaranteed to work)
plt.figure(figsize=(10, 6))
sns.heatmap(school_numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Across School Variables")
save_fig("school_numeric_correlation_heatmap")
plt.show()


# Select School Cost Columns
school_cost_df = school_data[
    ["school_monthly_water_cost_ssp", "school_monthly_water_cost_usd"]
].copy()

# Convert to numeric (force errors to NaN)
for col in school_cost_df.columns:
    school_cost_df[col] = pd.to_numeric(school_cost_df[col], errors="coerce")

# Remove Outliers Using IQR
Q1 = school_cost_df["school_monthly_water_cost_ssp"].quantile(0.25)
Q3 = school_cost_df["school_monthly_water_cost_ssp"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

school_cost_clean = school_cost_df[
    (school_cost_df["school_monthly_water_cost_ssp"] >= lower_bound)
    & (school_cost_df["school_monthly_water_cost_ssp"] <= upper_bound)
]


# -----------------------------------
# Boxplot After Cleaning (SSP)
# -----------------------------------
plt.figure(figsize=(8, 4))
sns.boxplot(x=school_cost_clean["school_monthly_water_cost_ssp"], color="teal")
plt.title("Distribution of School Monthly Water Costs (SSP)")
plt.xlabel("Cost (SSP)")
save_fig("school_monthly_water_cost_ssp")
plt.show()


#  Histogram of Cleaned SSP Values
plt.figure(figsize=(8, 4))
sns.histplot(
    school_cost_clean["school_monthly_water_cost_ssp"],
    bins=15,
    kde=True,
    color="purple",
)
plt.title("Histogram of Cleaned Monthly Water Costs (SSP)")
plt.xlabel("Cost (SSP)")
plt.ylabel("Frequency")
save_fig("school_monthly_water_cost_ssp_histogram")
plt.show()


#  Histogram of USD Values
plt.figure(figsize=(8, 4))
sns.histplot(
    school_cost_clean["school_monthly_water_cost_usd"],
    bins=15,
    kde=True,
    color="orange",
)
plt.title("Histogram of Monthly Water Costs (USD)")
plt.xlabel("Cost (USD)")
plt.ylabel("Frequency")
save_fig("school_monthly_water_cost_usd_histogram")
plt.show()


print("\n‚úî School exploration completed successfully!")
```
